```python
import numpy as np
import matplotlib.pyplot as plt

def plot_data_and_percentiles(data):
    percentiles = np.percentile(data, np.arange(101))
    
    highlight_points = [20, 30, 40, 50, 60]
    highlight_values = np.percentile(data, highlight_points)
    
    plt.figure(figsize=(12, 5))
    
    # Bar chart of the data
    plt.subplot(1, 2, 1)
    bars = plt.bar(range(len(data)), data, color='skyblue')
    plt.title('Reactions to Original Casts')
    plt.xlabel('Days Ago')
    plt.ylabel('Unique Users Reacting')
    
    # Annotate each bar with its data value
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), va='bottom')  # Adjusts vertical alignment
    
    # Distribution plot of percentiles
    plt.subplot(1, 2, 2)
    plt.plot(np.arange(101), percentiles, linestyle='-', color='skyblue', label='Percentiles')
    plt.scatter(highlight_points, highlight_values, color='green', marker='X', s=10, label='Highlighted Points')
    
    # Annotate highlighted percentile points
    for i, value in enumerate(highlight_values):
        plt.text(highlight_points[i], value, f"{round(value, 1)}", ha='right', va='bottom')
        
    plt.title('Approximate Percentiles')
    plt.xlabel('Percentile')
    plt.ylabel('Unique Users Reacting')
    plt.grid(True)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Sample data
data = [0, 45, 12, 34, 56, 78, 90, 100]

# Function call
plot_data_and_percentiles(data)
```
